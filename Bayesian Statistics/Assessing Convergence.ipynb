# Trace Plots and Autocorrelation

Before using simulated chains to provide Monte Carlo estimates, we should make sure that our chain has indeed converged to the stationary distribution.

Ensuring convergence is difficult but we can do several things to investigate.

## Autocorrelation

Autocorrelation is a number between -1 and 1 which measures how linearly dependent current value of the chain is to past values, also called lags.

$ACF(X_t, X_{t-l}) = \frac{Cov(X_t,X_{t-l})}{\sqrt{Var(X_t)  Var(X_{t-l})}}$

## Burn-In Period

Initial few iterations of the chain during which the parameter value goes from initial value to the stationary distribution.

One should always disregard first few points from the simulation results as these corresponds to the burn-in period of the simulation and do not correspond to the stationary distribution.

## Multiple Chains & Gelman-Rubin Diagnostics

One can run many chains simultaneously with differing initial values, but may notice that the chain converges nonetheless.

Gelman Rubin Daignostics takes in a list of chains and then figure out the variability within chain and variability amongst the chains

If the variability of chains (within a chain and amongst the chains) remains close to 1, that implies the chains have indeed reached the stationary distribution.
