```r
N = 1000 #N is the number of trials, i.e. coin-flips
```

```r
outcomes = sample(0:1, N, replace = TRUE)  #Outcome = 1 if it's HEADS
```

```r
outcomes
```

### Outcome = 1 if the event ended in the coin being Heads Upwards

```r
count = 1
running_proportion <- vector()
running_values = cumsum(outcomes)
for (running_value in running_values)
    {
    running_proportion <- append(running_proportion,(running_value/count))
    count = count + 1
}
```

```r
pdf("Running_proportion.pdf", width = 15,height = 15)
plot(running_proportion,type = 'o' )
xlim = c(1,1000)
ylim = c(0.00,1.00)
xaxs = 'i'
yaxs = 'i'
dev.off()
```

```r
xlow = -0.6
xhigh = 0.6
mean = 0.0
dev = 0.2
```

```r
x = seq(xlow,xhigh,by = 0.001)
```

```r
length(x)
```

```r
y = (1/(sqrt(2*pi) * dev)) * exp(-0.5*((x - mean) / dev)^2)
```

```r
i = seq(1,1201,50)
```

```r
x_i = x[i]
y_i = y[i]
```

```r
pdf("figure_2.pdf", width = 15,height = 15)
plot(x_i,y_i,type = 'h')
lines(x,y)
xaxs = 'i'
yaxs = 'i'
dev.off()
```
